#### Plotting Categorical Data
##### Jitter
##### Hue
##### Boxplot
##### Voilin Plot
##### Pointplot

```python
auto_data=pd.read_csv('Automobile.CSV')
```

```python
auto_data.head(10)
```

```python
sns.catplot(x="make", y="normalized_losses", data=auto_data)
```

```python
sns.catplot(x="make", y="normalized_losses", data=auto_data,jitter=False,height=10,aspect=2)
```

```python
sns.catplot(x="make", y="normalized_losses", data=auto_data,jitter=False,height=10,aspect=2,hue="drive_wheels")
```

## Boxplot using seaborn

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "box", data=auto_data)
```

### violin plot

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "violin", data=auto_data)
```

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "violin",split=False,hue='number_of_doors', data=auto_data)
```

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "bar",hue='number_of_doors', data=auto_data)
```

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "point",hue='number_of_doors', data=auto_data)
```

```python
sns.catplot(x="drive_wheels", y="peak_rpm", kind = "swarm",hue='number_of_doors', data=auto_data)
```

```python
sns.distplot(auto_data.engine_size)
```

```python
sns.distplot(auto_data.engine_size,kde=False,rug=True)
```
